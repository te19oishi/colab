```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
#import cis
```

```python
fs = 8000
t=np.arange(0,1,1/fs)
cs=np.exp(2j*np.pi*440*t)
y=np.real(cs)
plt.plot(y[:100])
plt.show()
```

```python
fs = 8000
t=np.arange(0,1,1/fs)
cs=np.exp(2j*np.pi*440*t)
y=np.imag(cs)
plt.plot(y[:100])
plt.show()
```

```python
fs = 8000
t=np.arange(0,1,1/fs)
cs=(np.exp(2j*np.pi*440*t)+np.exp(-2j*np.pi*440*t))/2
plt.plot(cs[:100])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import art3d

nend=63; n=np.arange(0, nend+1)
#W=np.pi/8
w=np.pi/9
ejen=np.exp(1j*w*n)
coswn=np.real(ejen)
sinwn=np.imag(ejen)

#
fig1=plt.figure()
ax1=fig1.add_subplot(1,1,1)
ax1.stem(n, coswn, markerfmt='x')
ax1.set_xlim(0,nend); ax1.set_ylim(-2,2)
ax1.grid()
ax1.set_xlabel('Time $n$')
ax1.set_ylabel('Real Part')
```

```python
#
fig2=plt.figure()
ax2=fig2.add_subplot(1,1,1)
ax2.stem(n, sinwn, markerfmt='.')
ax2.set_xlim(0, nend); ax1.set_ylim(-2, 2)
ax2.grid()
ax2.set_xlabel('Time $n$')
ax2.set_ylabel('Real Part')
```

```python
#
fig3=plt.figure()
ax3=fig3.add_subplot(1,1,1)
ax3.plot(coswn, sinwn, 'o')
for m in n:
  ax3.plot(np.array([0, coswn[m]]), np.array([0, sinwn[m]]))
ax3.axis('square')
ax3.set_xlim(-2, 2); ax1.set_ylim(-2, 2)
ax3.grid()
ax3.set_xlabel('Time $n$')
ax3.set_ylabel('Imaginary Part')
```

```python
fig4=plt.figure()
ax4=Axes3D(fig4)
ax4.plot(n, coswn, sinwn, 'o', label='$e^[j\omega n]$')

rmax=3
rplane= -rmax*np.ones(len(n))
iplane= -rplane
ax4.plot(n, coswn, rplane, 'x', label='$\cos \ \omega n$')
ax4.plot(n, iplane, sinwn, '.', label='$\sin \ \omega n$')
ax4.view_init(elev=35, azim=-50)
ax4.set_xlim(0, nend); ax4.set_ylim(-rmax,rmax)
ax4.set_zlim(-rmax, rmax); ax4.grid()
ax4.set_xlabel('Time n')
ax4.set_ylabel('Real Part')
ax4.set_zlabel('Imaginary Part')
ax4.legend(loc='upper right')
```
