```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread('paprika-966290_640.jpg')
plt.imshow(img_bgr)
```

```python
img_rgb=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_hls=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HLS_FULL)
img_hls
```

```python
C=cv2.imread('paprika-966290_640.jpg')
C_rgb=cv2.cvtColor(C,cv2.COLOR_BGR2RGB)
C=cv2.cvtColor(C,cv2.COLOR_BGR2HLS_FULL)#plt.imshow(C)#plt.show()
BW=np.zeros((C.shape[0],C.shape[1]))
BW[np.logical_and(np.logical_and(np.logical_and(C[:,:,0]>25,C[:,:,0]<45),    # 25<H<45
np.logical_and(C[:,:,1]>=0,C[:,:,1]<=255)), # 0<=L<=255
C[:,:,2]<=255)]=1                           # 0<=S<=255
R=C_rgb.copy()
R[BW==0]=0
plt.imshow(R)
plt.show()
```
